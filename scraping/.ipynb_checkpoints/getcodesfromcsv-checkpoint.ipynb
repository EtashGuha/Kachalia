```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
table = pd.read_excel('hopeful.xlsx')
table
```

```python
all_codes = []
for item in table["Unnamed: 0"]:
    all_codes.append(item.replace("-", ""))
```

```python
code_map = {}
```

```python
for code in all_codes[0:5]:
    information = []
    url = "https://icdlist.com/icd-10/%s" % code
    print(url)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    id_name = "Synonyms-%s" % code
    print(id_name)
    if soup.find(id=id_name):
        for item in list(soup.find(id=id_name).children):
            if item.name == 'ul':
                information.append(item.get_text(", "))
    gen_info = "Information-for-Patients-%s" % code
    if soup.find(id=gen_info):
        information.append(soup.find(id=gen_info).get_text(" "))
    code_map[code] = information
```
