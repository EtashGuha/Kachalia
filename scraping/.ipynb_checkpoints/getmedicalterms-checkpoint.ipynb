```python
import string
alphabet = list(string.ascii_lowercase)
complete_set = []
for letter in alphabet:
    url = "https://www.medicinenet.com/script/main/alphaidx.asp?p=" + letter + "_dict"
    print(url)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    all_terms = list(soup.find_all("li"))
    for term in all_terms:
        if "</a></li>" in str(term) and "<li><a href=" in str(term) and "onclick" not in str(term) and "target" not in str(term) and "alphaidx" not in str(term):
            new_term = term.get_text()
            if ", " in new_term:
                separated = new_term.split(", ")
                new_term = separated[1] + " " + separated[0]
            complete_set.append(new_term.lower())
```

```python
import regex as re
links = []
page = requests.get("https://www.icd10data.com/ICD10CM/Codes")
soup = BeautifulSoup(page.content, 'html.parser')
def remove_duplicates(l): # remove duplicates and unURL string
    for item in l:
        match = re.search("(?P<url>https?://[^\s]+)", item)
        if match is not None:
            links.append((match.group("url")))

data = []
for link in soup.find_all('a', href=True):
    data.append(str(link.get('href')))
flag = True
remove_duplicates(data)
print(data)
while flag:
    
    try:
        for link in links:
            for j in soup.find_all('a', href=True):
                temp = []
                source_code = requests.get(link)
                soup = BeautifulSoup(source_code.content, 'lxml')
                temp.append(str(j.get('href')))
                remove_duplicates(temp)

                if len(links) > 162: # set limitation to number of URLs
                    break
            if len(links) > 162:
                break
        if len(links) > 162:
            break
    except Exception as e:
        print(e)
        if len(links) > 162:
            break

for url in links:
    print(url)
```

```python

import requests 
from bs4 import BeautifulSoup 
from collections import Counter  
  
url = 'https://www.icd10data.com/ICD10CM/Codes/'
reqs = requests.get(url) 
print(reqs)
soup = BeautifulSoup(reqs.text, 'html.parser') 

def getNumberOfSlashes(stri):
    num = 0
    for i in stri:
        if i == "/":
            num += 1
    return num
urls = [] 
total_urls = []
print(getNumberOfSlashes("chicken/asb/tew"))
for link in soup.find_all('a'): 
    if getNumberOfSlashes(str(link.get('href'))) == 4 and "Rules" not in str(link.get('href')) and  "Changes" not in str(link.get('href')):
        total_urls.append(link.get('href'))

print(total_urls)
```

```python
final_urls = []
for end in total_urls:
    url = 'https://www.icd10data.com' + end
    reqs = requests.get(url) 
    print(reqs.status_code)
    soup = BeautifulSoup(reqs.text, 'html.parser') 
    for link in soup.find_all('a'): 
        if end in str(link.get('href')) and getNumberOfSlashes(str(link.get('href'))) >= 5 and "Rules" not in str(link.get('href')) and  "Changes" not in str(link.get('href')):
            final_urls.append(link.get('href'))
print(final_urls)
```
