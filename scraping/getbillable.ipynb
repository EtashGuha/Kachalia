```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
table = pd.read_excel('hopeful.xlsx')
table.tail()
```

```python
all_codes = []
for item in table["Unnamed: 0"]:
    all_codes.append(item.replace("-", ""))
```

```python
all_billings = {}
```

```python
for code in all_codes:
    try:
        if code in all_billings:
            continue
        billing_status = "Unknown"
        url = "https://icdlist.com/icd-10/%s" % code
        print(url)
        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser')
        stripped_code = code.replace(".", "")
        id_name = "icd-10-%s" % stripped_code
        print(id_name)
        if soup.find(id=id_name):
            if soup.find(id="smart-tags"):
                text = soup.find(id="smart-tags").get_text()
                if "Non-Billable" in text:
                    billing_status = "Billable"
                elif "Billable" in text:
                    billing_status = "Non-Billable"
        all_billings[code] = billing_status
    except:
        pass
```
