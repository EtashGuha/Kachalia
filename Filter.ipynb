```python
import pickle
from tqdm import tqdm
tqdm.monitor_interval = 0
icdToInfo = pickle.load(open("info.p", "rb"))
complete_set = pickle.load(open("complete_set.p", "rb"))
```

```python
print(complete_set)
```

```python
for blak in complete_set:
    if "(" in blak:
        complete_set.append(blak[0:blak.index("(") - 1])
        complete_set.append(blak[blak.index("(") + 1:-1])
complete_set = set(complete_set)
to_delete = []
for i in complete_set:
    if len(i) <= 2:
        to_delete.append(i)
        
for i in to_delete:
    complete_set.discard(i)
```

```python
complete_set = set(complete_set)
```

```python
import re 
count = 0
for icd in icdToInfo:
    text = icdToInfo[icd]["text"]
    count += 1
    if text is not None: 
        text = text.lower()
        keywords = []
        res = re.findall(r'\w+', text)
        
        for word in complete_set:
            if " " in word and word in text:
                keywords.append(word)
            else:
                if word in res:
                    keywords.append(word)
        icdToInfo[icd]["keywords"] = keywords
    if count % 1000 == 0:
        print(count)
```

```python
for i in icdToInfo:
    key = i.replace('-', '')
    for j in icdToInfo:
        if key in j and "keywords" not in icdToInfo[j] and "keywords" in icdToInfo[i]:
            icdToInfo[j]["keywords"] = icdToInfo[i]["keywords"]
```

```python
import pandas as pd

df = pd.DataFrame(data=icdToInfo)
df = df.T
df.to_excel("hopeful.xlsx")
```
